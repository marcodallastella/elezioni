```python
import pandas as pd
```

```python
df = pd.read_csv('../output/risultati_er.csv')
```

```python
df.columns
```

```python
# Create a dictionary mapping lists to candidates
candidate_coalitions = {
    'De Pascale Michele': [
        'RIFORMISTI EMILIA-ROMAGNA FUTURA - DE PASCALE PRESIDENTE', 
        'PARTITO DEMOCRATICO - DE PASCALE PRESIDENTE',
        'CIVICI, CON DE PASCALE PRESIDENTE',
        'ALLEANZA VERDI E SINISTRA - COALIZIONI CIVICHE - POSSIBILE',
        'MOVIMENTO 5 STELLE'
    ],
    'Elena Ugolini': [
        "FRATELLI D'ITALIA - GIORGIA MELONI",
        'RETE CIVICA - ELENA UGOLINI PRESIDENTE',
        'FORZA ITALIA - BERLUSCONI - UGOLINI PRESIDENTE - NOI MODERATI',
        'LEGA SALVINI EMILIA-ROMAGNA - IL POPOLO DELLA FAMIGLIA'
    ],
    'Luca Todori': ["LEALTA' COERENZA VERITA' - LUCA TEODORI"],
    'Federico Serra': ["EMILIA-ROMAGNA PER LA PACE, L'AMBIENTE E IL LAVORO"]
}

# Create new columns for each candidate's total votes
for candidate, lists in candidate_coalitions.items():
    # Sum up votes from all supporting lists
    df[f'{candidate}_total'] = df[lists].sum(axis=1)
    print(f"Calculated total votes for {candidate}")

# Print summary
print("\nSample of results:")
print(df[['comune'] + [f'{candidate}_total' for candidate in candidate_coalitions.keys()]].head())

# Calculate winning candidate for each comune
df['winning_candidate'] = df[[f'{candidate}_total' for candidate in candidate_coalitions.keys()]].idxmax(axis=1)
df['winning_candidate'] = df['winning_candidate'].str.replace('_total', '')

# Print summary statistics
print("\nOverall results:")
for candidate in candidate_coalitions.keys():
    total_votes = df[f'{candidate}_total'].sum()
    comuni_won = (df['winning_candidate'] == candidate).sum()
    print(f"\n{candidate}:")
    print(f"Total votes: {total_votes:,}")
    print(f"Comuni won: {comuni_won}")
```

```python
# Remove rows where 'comune' is either 'Sassofeltrio' or 'Montecopiolo'
df = df[~df['comune'].isin(['Sassofeltrio', 'Montecopiolo'])]

# Use .str.replace() to handle string replacement more safely
df['comune'] = df['comune'].str.replace("Forli'", "Forlì", regex=False)
df['comune'] = df['comune'].str.replace("Montescudo - Monte Colombo", "Montescudo-Monte Colombo", regex=False)

df.to_csv('../output/viz/risultati_er.csv', index=False, encoding='UTF-8-SIG')
```

```python
df['perc_diff'] = round(((df['De Pascale Michele_total'] - df['Elena Ugolini_total']) / df['Elena Ugolini_total']) * 100, 2)

df.to_csv('../output/viz/risultati_er.csv', index=False, encoding='UTF-8-SIG')
```
