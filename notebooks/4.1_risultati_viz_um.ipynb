```python
import pandas as pd
```

```python
df = pd.read_csv('../output/risultati_um.csv')
df.columns
```

```python
# Create a dictionary mapping lists to candidates
candidate_coalitions = {
    'Paolone Giuseppe': [
        'FORZA DEL POPOLO'
    ],
    'Marco Rizzo': [
        "ALTERNATIVA RIFORMISTA RIZZO PRESIDENTE",
        'DEMOCRAZIA SOVRANA POPOLARE'
    ],
    'Martina Leonardi': ["INSIEME PER UN'UMBRIA RESISTENTE"],
    'Elia Francesco Fiorini': ["ALTERNATIVA PER L'UMBRIA"],
    'Giuseppe Tritto': ["TRITTO PRESIDENTE UMANI INSIEME LIBERI"],
    'Donatella Tesei': [
        "UNIONE DI CENTRO",
        "ALTERNATIVA POPOLARE",
        "LEGA",
        "NOI MODERATI - CIVICI PER L'UMBRIA",
        "TESEI PRESIDENTE",
        "FORZA ITALIA",
        "FRATELLI D'ITALIA"
    ],
    'Stefania Proietti': [
        "UMBRIA DOMANI PROIETTI PRESIDENTE",
        "CIVICI UMBRI",
        "UMBRIA PER LA SANITA' PUBBLICA",
        "PARTITO DEMOCRATICO",
        "UMBRIA FUTURA - RIFORMISTI E CIVICI PROIETTI PRESIDENTE",
        "MOVIMENTO 5 STELLE",
        "ALLEANZA VERDI E SINISTRA"
    ],
    'Moreno Pasquinelli': ["FRONTE DEL DISSENSO"],
    'Fabrizio Pignalberi': [
        "QUINTO POLO PER L'ITALIA",
        "PIU' ITALIA SOVRANA"
    ]
}
```

```python
# Create new columns for each candidate's total votes
for candidate, lists in candidate_coalitions.items():
    # Sum up votes from all supporting lists
    df[f'{candidate}_total'] = df[lists].sum(axis=1)
    print(f"Calculated total votes for {candidate}")

# Print summary
print("\nSample of results:")
print(df[['comune'] + [f'{candidate}_total' for candidate in candidate_coalitions.keys()]].head())

# Calculate winning candidate for each comune
df['winning_candidate'] = df[[f'{candidate}_total' for candidate in candidate_coalitions.keys()]].idxmax(axis=1)
df['winning_candidate'] = df['winning_candidate'].str.replace('_total', '')

# Print summary statistics
print("\nOverall results:")
for candidate in candidate_coalitions.keys():
    total_votes = df[f'{candidate}_total'].sum()
    comuni_won = (df['winning_candidate'] == candidate).sum()
    print(f"\n{candidate}:")
    print(f"Total votes: {total_votes:,}")
    print(f"Comuni won: {comuni_won}")
```

```python
# Use .str.replace() to handle string replacement more safely
df['comune'] = df['comune'].str.replace("Citta'", "Città", regex=False)

df.to_csv('../output/viz/risultati_um.csv', index=False, encoding='UTF-8-SIG')
```
